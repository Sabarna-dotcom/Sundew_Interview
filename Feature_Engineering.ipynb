```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.feature_selection import f_regression
import joblib
```

```python
df = pd.read_csv("Data.csv")
```

```python
df.head()
```

```python
df = df.dropna()
```

```python
# Converting cholesterol level ranges to their midpoints

def range_to_mid(value):
    try:
        nums = [int(x) for x in value.split() if x.isdigit()]
        return np.mean(nums)
    except:
        return np.nan

df['cholesterol_level'] = df['cholesterol_level'].apply(range_to_mid)
```

```python
df['cholesterol_level'] = df['cholesterol_level'].astype(float)
```

```python
df.head()
```

```python
print(df.columns)
```

```python
df = df.drop(columns=["applicant_id","Location"], errors="ignore")
```

```python
print(df.columns)
```

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(f"{col} → {df[col].nunique()} unique values")
```

```python
cat_cols = df.select_dtypes(include='object').columns
le_dict = {}       

for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    le_dict[col] = le  
print("Label Encoding Completed")

joblib.dump(le_dict, "label_encoders.pkl")
```

```python
df.head()
```

```python
df.to_csv("insurance_data.csv", index=False)
```

```python

# Get all numeric columns AFTER encoding
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

# Remove the target column from feature list
numeric_cols = numeric_cols.drop("insurance_cost")

# Define X and y
X = df[numeric_cols]
y = df["insurance_cost"]

# Run ANOVA F-test
f_scores, p_values = f_regression(X, y)

# Results
anova_results = pd.DataFrame({
    "Feature": numeric_cols,
    "F_score": f_scores,
    "p_value": p_values
}).sort_values("F_score", ascending=False)

anova_results
```

```python
selected_features = anova_results[
    (anova_results["p_value"] < 0.05)
].sort_values("F_score", ascending=False)

removed_features = anova_results[
    (anova_results["p_value"] >= 0.05)
].sort_values("F_score")
```

```python
print("Features to KEEP:")
print(selected_features)

print("\nFeatures to REMOVE:")
print(removed_features)
```
